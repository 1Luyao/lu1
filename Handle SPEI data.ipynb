# Handle SPEI data


```python
import pandas as pd
file_paths = [
    'SPEI Data/110.250000_35.250000_spei1.csv', 
    'SPEI Data/110.250000_35.250000_spei3.csv', 
    'SPEI Data/110.250000_35.250000_spei6.csv', 
    'SPEI Data/110.250000_35.250000_spei12.csv'   
]
first_df = pd.read_csv(file_paths[0])
first_split = first_df[first_df.columns[0]].str.split(';', expand=True)
first_split.columns = ['date', first_df.columns[0].split(';')[1]]
merged_df = first_split
for file_path in file_paths[1:]:
    df = pd.read_csv(file_path)
    split_df = df[df.columns[0]].str.split(';', expand=True)
    split_df.columns = ['date', df.columns[0].split(';')[1]]
    merged_df = pd.merge(merged_df, split_df, on='date', how='outer')

merged_df = merged_df.sort_values(by='date')

csv_path = 'spei_original.csv'
merged_df.to_csv(csv_path, index=False)
```
